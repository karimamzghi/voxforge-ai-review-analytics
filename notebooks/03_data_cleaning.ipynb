```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent
sys.path.append(str(PROJECT_ROOT))
```

```python
import pandas as pd

from src.voxforge.data.clean import clean_dataset
from src.voxforge.data.config import (
    INTERIM_DATA_DIR,
    PROCESSED_DATA_DIR,
)
```

```python
# load the interim merged dataset
merged_path = INTERIM_DATA_DIR / "merged_reviews.csv"

merged_df = pd.read_csv(
    merged_path,
    low_memory=False,
)
```

```python
# apply the cleaning rules inside clean_dataset method to the merged dataset
cleaned_df = clean_dataset(merged_df)
```

```python
# compare merged dataset before and after cleaning
print("Before cleaning:", merged_df.shape)
print("After cleaning:", cleaned_df.shape)
print("Rows removed:", len(merged_df) - len(cleaned_df))
```

```python
# analyse the cleaned dataset
cleaned_df.info(memory_usage="deep")
```

```python
# check for empty objects in the cleaned dataset
cleaned_df.isna().sum().sort_values(ascending=False)
```

```python
# save the cleaned dataset
PROCESSED_DATA_DIR.mkdir(
    parents=True,
    exist_ok=True,
)

cleaned_path = (
    PROCESSED_DATA_DIR
    / "cleaned_reviews.csv"
)

cleaned_df.to_csv(
    cleaned_path,
    index=False,
)

print(f"Saved cleaned dataset to: {cleaned_path}")
```
